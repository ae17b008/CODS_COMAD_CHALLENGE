```python
import pandas as pd
import numpy as np
import os
```

```python
data_files  = ['1', '2', '3', '4', '5', '6', '8', '9', '10', '11']
preds = []
```

```python
for file in data_files:
    df = pd.read_csv(os.path.join('data', f"{file}.csv"))
    df['label'] = df['label'].astype(int)
    preds.append(df['label'].values)
    
```

```python
preds = np.array(preds)
```

```python
from scipy import stats
preds = stats.mode(preds, axis=0).mode[0]
```

```python
preds
```

```python
# create a submission file
sub = pd.DataFrame({'ID': df['ID'], 'label': preds})
```

```python
# save submission file
sub.to_csv(os.path.join('..', 'output', 'voting.csv'), index=False)
```
